```python
distributions = []
indx = 0
for subcorpus in subcorpora:
    frequencies = list(subcorpus.values())
    relative_freq = [round(frequency/num_texts[indx], 2) for frequency in frequencies ]
    dist = dict(Counter(relative_freq))
    distributions.append(dist)
    indx += 1
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.style.use('ggplot')

dist = distributions[0]
plt.bar(list(dist.keys()), dist.values(), color='b', width=0.005)
plt.xticks(np.arange(0, 0.1, 0.01))
plt.xlim(0, 0.1)
plt.title('Распределение частотности тем, 2003-2008')
plt.xlabel('Доля темы в подкорпусе')
plt.ylabel('Число тем')
plt.show()

# Density
s = pd.Series(dist)
s.plot.kde()
```
